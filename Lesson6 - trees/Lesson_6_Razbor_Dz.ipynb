## Домашнее задание: Kaggle. Титаник.

```python
import pandas as pd
data_train = pd.read_csv('data/titanic/train.csv')
data_train.head()
```

```python
data_test = pd.read_csv('data/titanic/test.csv')
data_test.head()
```

## Предобработка данных

### Задание 1. Закодировать категориальные признаки числами (2 балла)
Пример: в колонке Sex сейчас значения male и female. Вы должны все male заменить на 1, все female - на 2. Аналогично для Embarked

```python
data_train.loc[data_train.Sex == 'male','Sex'] = 1
data_train.loc[data_train.Sex == 'female','Sex'] = 2

data_train.loc[data_train.Embarked == 'S','Embarked'] = 1
data_train.loc[data_train.Embarked == 'C','Embarked'] = 2
data_train.loc[data_train.Embarked == 'Q','Embarked'] = 3

data_train.head()
```

```python
data_test.loc[data_test.Sex == 'male','Sex'] = 1
data_test.loc[data_test.Sex == 'female','Sex'] = 2

data_test.loc[data_test.Embarked == 'S','Embarked'] = 1
data_test.loc[data_test.Embarked == 'C','Embarked'] = 2
data_test.loc[data_test.Embarked == 'Q','Embarked'] = 3

data_test.head()
```

### Задание 2. Заменить пропущенные значения Age и Fare на медианные значения (2 балла)
Если кто-то забыл что такое медиана - гугл вам поможет

```python
data_train.loc[data_train.Age.isnull(),'Age'] = data_train['Age'].median()
data_train.loc[data_train.Fare.isnull(),'Fare'] = data_train['Fare'].median()

data_test.loc[data_test.Age.isnull(),'Age'] = data_test['Age'].median()
data_test.loc[data_test.Fare.isnull(),'Fare'] = data_test['Fare'].median()
```

### Задание 3. Записать признаки, которые вы будете использовать в переменную X, а целевую переменную в y (2 балла)
Мы делали похожее задание на уроке, если не знаете как делать - смотрите конспект.

```python
data_train.head()
```

```python
X = data_train.ix[:, [2, 4, 5, 6, 7, 9]].values
y = data_train.ix[:, 1]
```

### Задание 4. Обучить модель логистической регрессии, сделать предсказания для тестовой выборки, загрузить решение на kaggle (4 балла)

```python
from sklearn.linear_model import LogisticRegression
C = 1
logit = LogisticRegression(C=C, n_jobs=-1, random_state=17)
logit.fit(X, y)
round(logit.score(X, y), 3)
```

```python
X_test = data_test.ix[:, [1, 3, 4, 5, 6, 8]].values
y_test = logit.predict(X_test)
```

```python
with open('prediction.txt', 'w') as f:
    for i in y_test:
        f.write(str(i) + '\n')
```
